```python
# Git- A version control system which is popular, free & opensource, fast & scalable
# A version control system is a tool that helps to track changes in code
# Github - Website that allows developers to store and manage their code using Git
# Each folder in github is called a repository
# When creating a new repository, name the repository by seperating words with a dash(-) and always remember to add a README file(description about project) since it helps in easy initialization of repo
# A Git commit is like taking a snapshot of your project at a specific point in time.

# Configuring git - telling git about which account are we making changes to
# there are 2 ways in configuring - local and global
# GLOBAL- these settings apply to all your git repositories on your machine
git config --global user.name "My Name"
git config --global user.email "someone@email.com"

# LOCAL - these settings are specific to a particular Git repository.(use cd to go that particular repo)
git config user.name "My Name"
git config user.email "someone@email.com"

# to list the changes that we made using configuration
git config --list

# Clone - cloning a repository on our local machine from some remote repo (drop down of green code button has the https link of the repo in github)
git clone <-some link->

# whenever git tracks a folder or repository, it creates a .git folder by default inside the repo
# to check the status of our code (tells you whether u are ahead or behind relative to the main branch)
git status

# when you modify a file inside VS code, the git extensions shows you symbols to the right side of file
# Thare are 4 status
# untracked - new files that git doesnt yet track (U)
# modified - changed (M)
# staged - file is ready to be committed (A)
# unmodified - unchanged

# whenever a file is modified or new files are added - they are in unstaged mode
# git add . -> stages the changes of all files (stages meaning that they are ready to be commited)
# git add <-file name->
# add - adds new or changed files in your working directory to the Git staging area

git add .
# This command stages all files and directories in the current directory and its subdirectories for the next commit.
# It does not consider files that have been deleted.

git add -A
# This command stages all files, including deleted files (removing them from the staging area), modified files, and untracked files, in the entire working directory and its subdirectories for the next commit.
# It stages all changes, additions, and removals.

Both git add . and git add -A achieve similar results of staging all files in the current directory and its subdirectories. However, they have slightly different behaviors:

git add .:

This command stages all files and directories in the current directory and its subdirectories for the next commit.
It does not consider files that have been deleted.
git add -A:

This command stages all files, including deleted files (removing them from the staging area), modified files, and untracked files, in the entire working directory and its subdirectories for the next commit.
It stages all changes, additions, and removals.

# So, if you want to stage all changes, including deletions, modifications, and additions, git add -A is more appropriate.
# However, if you're only interested in staging modifications and additions, excluding deletions, git add . suffices.

# commit - it is record of change
git commit -m "some message"

# after commiting, when we use git status -> you are ahead of main branch by 1 commit. so to update we push the files
# git push - upload local repo content to remote repo
git push origin main  (even "git push" is enough)

# to get the updated repo in another machine use:
git pull

# init - used to create a new git repo
git init
# create a new folder by mkdir <-foldername->
# locate to the folder by cd and type git init  (initialized empty git repository and .git file is added)
# after adding some files to the new folder locally, we use git add -a and git commit -m "message"
# now to upload the local repo to github first we create a repo manually in github with the same name
# before pushing local repo to the newly created repo in github we use
git remode add origin <-link-> (the remote refers to repo)
# to verify what repo we have pushed
git remote -v (to veriy remote)

# a Git branch is like a separate line of development within a Git repository.
# It allows you to work on different aspects of your project simultaneously without affecting the main or other branches.
# Imagine it as a way to isolate your work on a specific feature or bug fix, making it easier to manage and merge changes later

# to check branch
git branch

# to rename branch
git branch -M newname

# after checking and renaming branches, u can use git push origin main to upload the newly created git repo
git push -u origin main -> setting upstream (once we type this command, it is enough to type git push next time onwards)
# -u basically tells us that we are gonna work on origin main for a long time and we dont neet to type origin main everytime then onwards

# branch commands
git branch (to check branch)
git branch -M main (to rename branch)
git checkout <-branch name-> (to navigate)
git checkout -b <-new branch name-> (to create new branch)
git branch -d <-branch name-> (to delete branch)

# lets say you are in main branch
# when you type git branch
* main
# when you type git checkout -b feature1
Switched to a new branch 'feature1'
# when you type git branch again
* feature1
  main
# when you type git checkout main
switched to branch 'main'
your branch is up to date with 'origin/main'
# type in git branch again
  feature1
* main

# git branch -b feature2
switched to a new branch 'feature2'

# typing git branch -d feature2 -> gives error since you cannot delete a branch without navigating out of it

# appropriate method of deleting:
# git checkout main
# git branch -d feature2
deleted branch feature2

# if you modify feature1 and type git push origin feature1
# in github there will be a pop up "feature1 had recent pushes less than a minute ago"
# beside the pop up - compare and pull request button

# merging code
# way 1
git diff <-branch name-> (to compare commits,branches,files & more)
git merge <-branch name-> (to merge 2 branches)

# way 2
create a PR (pull request)
# pull request -> it lets you tell others about changes you've pushed to a branch in a repository on Github
# the senior developer reviews the PR and can comment on changes

# click compare and pull request and click create a pull request(you can add the description)
# github can sense whether there is conflict in the PR. if not then it can merge the PR
# a conflict in Git happens when two different changes are made to the same part of a file, and Git doesn't know which change to accept.
# It's like when two people try to change the same sentence in a document differently, and you're not sure which version is correct.

# pull command
git pull origin main
# used to fetch and download content from a remote repo and immediately update the local repo to match that content

# resolving merge conflicts
# an event that takes place when git is unable to automatically resolve differences in code between two commits

# when git merge two such conflicted codes(lets say merging main onto feature1 and they both have different codes on same line) - we get resolve in merge editor option
# moreover there are options like:
# accept current change - feature1 alone
# accept incoming change - main alone
# accept both changes - then the content of main will be added after feature1

# git pull is a Git command used to fetch and merge changes from a remote repository into your local repository
# while a pull request is a feature provided by Git hosting platforms for proposing and discussing changes before they are merged into a repository.

# undoing changes
# case 1: staged changes (changes that are added but not committed)
# git reset <-file name->
# get reset

# case 2: commited changes(for one commit)
# git reset HEAD~1 (the latest commit is refered as HEAD. so HEAD~1 means take the HEAD one step back and making it as latest commit)

# case 3: commited changes(for many commits)
# git reset <-commit hash->
# git reset --hard <-commit hash->
# each commit has a unique hash value. to return to specific commits back, copy the hash and type the above commands
# git log shows the hash of each commit
# The git log command is used to display a history of commits in a Git repository. When you run git log in your terminal or command prompt within a Git repository,
# it shows you a chronological list of commits, starting from the most recent commit and going backwards in time.

# Git fork
# a fork is a new repository that shares code and visibility settings with the original "upstream" repository
# fork is a rough copy
# Use git clone to make a local copy of an existing repository onto your machine for development or collaboration.
# Use git fork to create a personal copy of a repository under your own account on a Git hosting platform, allowing you to modify it independently of the original repository.

# simply, git clone is used if you wanna work on your own project, git fork is used to work on others project
# the owner can view the persons who forked his repo and after forking they can further add some features and send a pull request for merging
# the owner can also limit the ability to open a pull request from new users

# somemore git commands
# 1. git stash: temporarily stores changes that are not ready for commit, allowing you to switch branches without committing them
# 2. git rebase: reapplies commits from one branch on top of another, which can provide a linear history
# 3. git fetch: retrives changes from a remote repository without merging them
# 4. git tag: lists,creates or deletes tags to mark specific commits
# 5. git blame: shows who last modified each line of a file, helping you identify the author of specific changes
# 6. git remote add/remove: adds/removes a remote repository to/from your git configuration.
```
